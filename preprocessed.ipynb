```python
df_test = df[df["Purchase"].isnull()]
```

```python
df_train = df[-df["Purchase"].isnull()]
df_train.head()
```

```python
X = df_train.drop("Purchase",axis = 1)
X.head()
```

```python
y = df_train["Purchase"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.drop('Product_ID',axis = 1,inplace = True)
X_test.drop('Product_ID',axis = 1,inplace = True)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=0.01),
    "Random Forest": RandomForestRegressor(n_estimators=300, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=300, learning_rate=0.05),
    "XGBoost": XGBRegressor(n_estimators=500, learning_rate=0.05, max_depth=6, random_state=42),
    "LightGBM": LGBMRegressor(n_estimators=500, learning_rate=0.05)}

results = []

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    results.append((name, rmse))
    print(f"{name}: RMSE = {rmse}")

# Convert results to DataFrame
results_df = pd.DataFrame(results, columns=["Model", "RMSE"]).sort_values(by="RMSE", ascending=True)

print("Best Models Ranked:")
print(results_df)

# Select best model
best_model_name = results_df.iloc[0, 0]
best_rmse = results_df.iloc[0, 1]
best_model = models[best_model_name]

print(f"\n Best Model: {best_model_name} with RMSE = {best_rmse}")
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=0.01),
    "Random Forest": RandomForestRegressor(n_estimators=300, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=300, learning_rate=0.05),
    "XGBoost": XGBRegressor(n_estimators=500, learning_rate=0.05, max_depth=6, random_state=42),
    "LightGBM": LGBMRegressor(n_estimators=500, learning_rate=0.05)}

results = []

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    # Handle zero values in y_test to prevent division by zero
    y_test_nonzero = np.where(y_test == 0, 1e-7, y_test)
    mape = mean_absolute_percentage_error(y_test_nonzero, y_pred)
    accuracy = 100 - mape * 100

    results.append((name, rmse, accuracy))
    print(f"{name}: RMSE = {rmse:.2f}, Accuracy = {accuracy:.2f}%")

# Create DataFrame with results
results_df = pd.DataFrame(results, columns=["Model", "RMSE", "Accuracy (%)"]).sort_values(by="Accuracy (%)", ascending=False)

print("\nBest Models Ranked by Accuracy:")
print(results_df)

# Get best model by accuracy
best_model_name = results_df.iloc[0]["Model"]
best_model = models[best_model_name]
best_accuracy = results_df.iloc[0]["Accuracy (%)"]

print(f"\nBest Model: {best_model_name} with Accuracy = {best_accuracy:.2f}%")
```
